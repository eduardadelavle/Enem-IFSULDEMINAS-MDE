# **TRANSFORMAÇÃO TIPO SEXO, ESCOLA E RAÇA**

Utilizando a planilha csv dados brutos iremos para a transformação dos dados. Portanto, os atributos das colunas sexo, escola e raça foram transformados em conjuntos de dados com o objetivo de facilitar seu uso pelas técnicas de mineração e enriquecer os resultados.

* Obs: criação de um dicionário - transformando os atributos do próprio dicionário do ENEM disponibilizado pelo INEP.

# **Boxplot**
Média das notas de todas as áreas do conhecimento

```
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(style="whitegrid", color_codes=True)
sns.despine()
plt.figure(figsize = (12, 8))
plt.xticks(rotation=35, ha ="right")
plt.xlabel('')
sns.boxplot(data=micronotas);
plt.savefig('boxplot.png', format='png')
```
